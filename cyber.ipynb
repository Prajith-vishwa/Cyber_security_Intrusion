# COMBINED DATASETS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load dataset
train_data = pd.read_csv("Train_data.csv")
test_data = pd.read_csv("Test_data.csv")
```

```python
# Combine datasets for EDA
data = pd.concat([train_data, test_data], axis=0)
```

```python
# Display dataset info
print("\nDataset Info:")
print(data.info())
```

```python
#Display first few rows
print("\nFirst 5 rows of the dataset:")
print(data.head())
```

```python
# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Drop rows with missing values
data.dropna(inplace=True)
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Check for duplicate rows
print("\nNumber of duplicate rows:", data.duplicated().sum())
```

```python
# Select numerical and categorical columns
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
cat_cols = data.select_dtypes(include=['object']).columns
```

#CLASS DISTRIBUTION

```python

if 'class' in data.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x='class', data=data, palette='coolwarm')
    plt.title("Class Distribution")
    plt.show()
else:
    print("\n⚠ Target variable 'class' not found! Check column names.")
```

#HISTOGRAM

```python

plt.figure(figsize=(12, 10))
data[num_cols].hist(bins=30, edgecolor='black', figsize=(12, 10))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

#CORRELATION HEATMAP

```python

plt.figure(figsize=(12, 8))
sns.heatmap(data[num_cols].corr(), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

#BOX PLOT

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=data[num_cols])
plt.xticks(rotation=90)
```

#VIOLIN PLOTS

```python

if 'class' in data.columns:
    plt.figure(figsize=(12, 6))
    for col in num_cols[:5]:  # Limit to first 5 features
        plt.figure(figsize=(6, 4))
        sns.violinplot(x='class', y=col, data=data, palette="coolwarm")
        plt.title(f"Violin Plot of {col} by Class")
        plt.show()
```

#KDE PLOT

```python

for col in num_cols[:5]:  # Limit to first 5 features
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data[col], fill=True, color="skyblue")
    plt.title(f"KDE Plot of {col}")
    plt.show()
```

#FEATURE IMPORTANCE


```python

if 'class' in data.columns:
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.preprocessing import LabelEncoder

    # Encode labels if necessary
    le = LabelEncoder()
    y = le.fit_transform(data['class'])

    # Train a quick RandomForest model for feature importance
    X = data[num_cols].drop(columns=['class'], errors='ignore')
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X, y)

    # Plot feature importance
    feature_importances = pd.Series(model.feature_importances_, index=X.columns)
    feature_importances.nlargest(15).plot(kind='barh', colormap='coolwarm')
    plt.title("Top 15 Feature Importances")
    plt.show()
```

#DETAILED CORRELATION

```python

plt.figure(figsize=(10, 6))
sns.heatmap(data[num_cols].corr(), annot=True, cmap='coolwarm', linewidths=2.5, fmt=".2f")
plt.title("Detailed Feature Correlation Heatmap")
plt.show()
```

#PIE CHART

```python
 
if 'class' in data.columns:
    plt.figure(figsize=(6, 6))
    data['class'].value_counts().plot.pie(autopct='%1.1f%%', cmap='coolwarm', startangle=90, wedgeprops={'edgecolor': 'black'})
    plt.title("Class Distribution")
    plt.ylabel("")  # Hide y-label for better visualization
    plt.show()
else:
    print("\n⚠ Target variable 'class' not found! Check column names.")
```
